# The Cost Function

---


The **cost function**, $C(q)$, gives the total cost of producing a quantity $q$ of some good. The more goods that are made, the higher the total cost, so $C$ is an increasing function. Costs of production can be separated into two parts:

$$\text{Total cost} = \text{Fixed costs} + \text{Variable costs}$$

- The **fixed costs** are incurred even if nothing is produced ($q=0)$.
  - The fixed costs are the vertical intercept.
  - Costs such as purchasing machinery and paying rent for a factory are examples of fixed costs since they do not depend on how many goods $q$ are produced.
- The **variable costs** depend on how many goods $q$ are produced.
  - Costs such as labor and raw materials are examples of variable costs.

# Symbolic Differentiation with `diff()`

---

To calculate the derivative of $f$ at several different values of $x$, it is often more efficient to first find a general formula for the derivative $f'$, and then evaluate the formula for $f'$ at different values of $x$.

The [`diff()`](https://docs.sympy.org/latest/tutorials/intro-tutorial/calculus.html#) function in SymPy finds a symbolic formula for the derivative of a symbolic function. To symbolically differentiate a function `y`, we can perform the following steps:

1. Define a symbol used for the input variable.
2. Define the symbolic function in terms of the symbol.
3. Differentiate using the command `sym.diff(symbolic_function, input_symbol)`.


## Example: Differentiating $f(x) = x^5 + 2x$ with `diff()`

---

To find the derivative of $f(x) = x^5 + 2x$:

1. Define a symbol used for the input variable: `x = sym.symbols('x')`.

2. Define the symbolic function: `y = x**5 + 2 * x`.

3. The command `sym.diff(y, x)` returns a formula for the derivative, $f'(x) = 5x^4 + 2$.



```python
########################################
# A method for symbolic differentiation
########################################
x = sym.symbols('x')  # define x as symbol
y = x**5 + 2 * x  # define symbolic function

y_prime = sym.diff(y, x)  # find the derivative of y with respect to x
y_prime
```

### Using `y.diff(x)` or `sym.diff(y, x)`

---

We can also call `diff()` as a [method](https://www.geeksforgeeks.org/difference-method-function-python/) by first entering the name of the symbolic function followed by `diff(input_symbol)`.

1. Define a symbol used for the input variable: `x = sym.symbols('x')`.

2. Define the symbolic function: `y = x**5 + 2 * x`.

3. The command `y.diff(x)` returns a formula for the derivative, $f'(x) = 5x^4 + 2$.

**The commands `sym.diff(y, x)` and `y.diff(x)` do the same exact operation. Choose either method!**

<br>  

Run the code cell below to verify that the output of the command `y.diff(x)` is identical to the output we obtained from `sym.diff(y, x)` in the previous code cell.



```python
##############################################
# Another method for symbolic differentiation
##############################################
x = sym.symbols('x')  # define x as symbol
y = x**5 + 2 * x  # define symbolic function

y_prime = y.diff(x)  # find the derivative of y with respect to x
y_prime
```

# Computing Higher-Order Derivatives

---

We have used Python to find a formula for the derivative of $y = x^5 + 2x$ which is $y' = 5x^4 + 2$.

The second derivative, $y''$, can be found by taking the derivative of $y'$ again,

$$y'' = \big( y' \big)' = \big( 5x^4 + 2 \big)' = 20 x^3.$$

In Python, we can find formulas for higher-order derivatives, such as $y''$ and $y'''$, without first finding a formula for the first derivative $y'$.  To find the n$^{\text{th}}$-order derivative of `y` with respect to `x`, we can add an optional argument inside `diff()` that specifies the order, `n`, of the derivative.

- For example, either `sym.diff(y, x, 2)` or `y.diff(x, 2)` can be used to find a formula for the second derivative, $y''$.

- For the first derivative, we do not need to specify any value for `n`. Both `y.diff(x)` and  `y.diff(x, 1)`  identically find a formula for $y'$.

**Experiment with the code cell below to find formulas for higher order-derivatives of $y=x^5 + 2x$.**




```python
################################
# Adjust the value of n to find
# higher-order derivatives.
################################
x = sym.symbols('x')  # define x as symbol
y = x**5 + 2 * x  # define symbolic function

sym.diff(y, x, 2)  # find higher-order derivatives
```

## Symbolic Differentiation with `diff()`

---

```python
# define the function
import sympy as sym
x=sym.symbols('x')
y = x**2 * sym.sin(1 + x**3) + sym.exp(2*x)

# derivative of y with respect to x
sym.diff(y, x)

#########################
# alternate notations:
########################
# y.diff(x)
# sym.diff(y) # <--- assumes x is the independent variable
# y.diff()    # <--- assumes x is the independent variable
```

This all works with different variables too. The code cell below is exactly like the one above, but it uses $T$ and $r$ instead of $y$ and $x$.

```python
# define the function
import sympy as sym
r=sym.symbols('r')
T = r**2 * sym.sin(1 + r**3) + sym.exp(2*r)

# derivative of T with respect to r
sym.diff(T, r)

#########################
# alternate notations:
########################
# T.diff(r)
# sym.diff(T) # <--- assumes r is the independent variable
# T.diff()    # <--- assumes r is the independent variable
```

## Higher-Order Derivatives

---


```python
# define the function
import sympy as sym
x=sym.symbols('x')
y = x**2 * sym.sin(1 + x**3) + sym.exp(2*x)

# 2nd order derivative of y with respect to x
sym.diff(y, x, 2)

#########################
# alternate notations:
########################
# y.diff(x,2)
```
