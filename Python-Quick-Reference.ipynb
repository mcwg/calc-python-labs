# Solving equations numerically with SciPy

```python
from scipy.optimize import fsolve
import numpy as np

# To solve an equation of the form f(t) = g(t), we rewrite it as f(t) - g(t) = 0
# and use "fsolve" to figure out when  f(t) - g(t) is zero

# Example:
# solve the equation 1.5e^(0.03 t) = 0.2t

def f(t): # left-hand-side of equation
  return 1.5 * np.exp(0.03 * t)

def g(t): # right-hand-side of equation
  return 0.2 * t

def equation(t):
  return f(t) - g(t)

initial_guess = 10  # need to provide initial guess

solutions = fsolve(equation, initial_guess)  # find solutions

print("Solution(s) at t =", solutions)

# check
print("value of f at the solution:", f(solutions[0]))
print("value of g at the solution:", g(solutions[0]))
```

## Compute a new column from values in other rows (and other columns)



```python
# create the dataframe for this example
import numpy as np
import pandas
def f(x):
  return 1 / ( 1 + x**2 )
x_values = np.linspace(-5,5,100)
f_values = f(x_values)
df = pandas.DataFrame({"x": x_values, "f(x)": f_values})

# Create a new column named 'derivative' which contains average rate of change between the top and bottom values of the 'y' column
df['derivative of f (approximate)'] = ( df['f(x)'].shift(1) - df['f(x)'].shift(-1) ) / ( df['x'].shift(1) - df['x'].shift(-1) )
df
```

```python
# create the dataframe for this example
import numpy as np
import pandas
def f(x):
  return 1 / ( 1 + x**2 )
def f_prime(x):
  return -2*x / ( 1 + x**2 )**2
x_values = np.linspace(-5,5,100)
f_values = f(x_values)
f_prime_values = f_prime(x_values)
df = pandas.DataFrame({"x": x_values, "f(x)": f_values, "f'(x) exact": f_prime_values})

# Create a new column named 'derivative' which contains average rate of change between the top and bottom values of the 'y' column
df['derivative of f (approximate)'] = ( df['f(x)'].shift(1) - df['f(x)'].shift(-1) ) / ( df['x'].shift(1) - df['x'].shift(-1) )
df
```

The following example shows how to extract values in a dataframe to compute an average rate of change.

```python
# load data for this example
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/mcwg/calc-python-labs/main/Data/CO2-emissions.csv')

# print de data
print(df)

####################################
# Compute average rate of change
####################################

# Extract start and end values of t
t0 = df.loc[4, 'Year']  # start time (year 2014 is index 4)
t1 = df.loc[5, 'Year']  # end time   (year 2015 is index 5)

# Extract corresponding start and end water level elevations
f0 = df.loc[4, 'CO2_Emissions']  # value of f at start time
f1 = df.loc[5, 'CO2_Emissions']  # value of f at end time

# Compute average rate of change to approximate f'(2014)
average_rate = ( f1 - f0 ) / ( t1 - t0 )
print("The average rate of change between t =",t0, "and t =",t1,"is", average_rate)
```

Alternatively, seach for the right rows in the dataframe and extract the values.

```python
# load data for this example
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/mcwg/calc-python-labs/main/Data/CO2-emissions.csv')

# print de data
print(df)

####################################
# Compute average rate of change
####################################

# Define the start and end years
start_year = 2011
end_year = 2018

# Get CO2 emissions for the start and end years
# .values[0] extracts the first (and only) element from the array of matching CO2 emissions
row_start_year=df[df['Year'] == start_year]
row_end_year=df[df['Year'] == end_year]
start_emission = row_start_year['CO2_Emissions'].values[0]  # Value of CO2_Emissions for start_year
end_emission = row_end_year['CO2_Emissions'].values[0]      # Value of CO2_Emissions for end_year

# Calculate the average rate of change per year
average_rate = (end_emission - start_emission) / (end_year - start_year)

# Print the result
print(f"The average rate of change in CO2 emissions per year between {start_year} and {end_year} is {average_rate}")
```

Since the average time it takes for all the blood to return to the heart is approximately the same for any mammal with the same mass, we can think if the circulation time, $T$, as a *function* of the mass, $m$, and write
$$T=f(m).$$

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Create dataframe for this example
data = {
    'mass': [70, 180, 700,1170,2520,3000,5450,91000],
    'Time': [50, 64, 90, 102, 123, 129, 150, 302]
}
mamal_data = pd.DataFrame(data)

# Create array of x values and y values for the model T= f(m) = 17.4*m^(1/4)
x_values = np.arange(0, 100000, 1) # Syntax is arange( start, end, step-size)
def f(m):
  return 17.4*m**(1/4)
y_values = f(x_values)

# Plot both data and model together
plt.plot(x_values, y_values, 'b-')
plt.plot(mamal_data['mass'], mamal_data['Time'], 'r', marker='o', linestyle='None')
```

- Use `sympy.sin(x)`, `sympy.cos(x)`, and `sympy.tan(x)`  for $\sin{x}$, $\cos{x}$, and $\tan{x}$, respectively.
- Use `sympy.log(x)` for $\ln{x}$.
- Recall in Python `log` typically means natural log and `log10` is used for log base 10.
- See the [SymPy manual](https://docs.sympy.org/latest/modules/functions/elementary.html) for a complete list of available mathematical functions.

You can replace `sympy.` by `sym.` if you import `sympy` as `sym`, like in the code cells below.

```python
# import sympy and declare x to be a variable
import sympy as sym
x=sym.symbols('x')

# Define y using the symbolic variable x
y = 2 * x**5 - 3 * sym.exp(x) - sym.cos(8*x)

# print y
y
```

## Evaluating a symbolic functions

```python
# Define the function
import sympy as sym
x=sym.symbols('x')
y = x**2 * sym.sin(1 + x**3) + sym.exp(2*x)

# evaluate at x=3
y.evalf(subs={x: 3}) # this evaluates y at x=3
```

Alternatively, use `subs` directly, but note that this gives an expression, not a decimal:

```python
# Define the function
import sympy as sym
x=sym.symbols('x')
y = x**2 * sym.sin(1 + x**3) + sym.exp(2*x)

# evaluate at x=3
y.subs(x,3)
```

To get a decimal value, use `float()` together with `subs()`

```python
# Define the function
import sympy as sym
x=sym.symbols('x')
y = x**2 * sym.sin(1 + x**3) + sym.exp(2*x)

# evaluate at x=3 and convert into a decimal
float(y.subs(x,3))
```

```python
# load libraries and define function for this cell to run
import sympy as sym
x = sym.symbols('x')
f = x**2

# To solve f(x) = 16 we need to rewrite as f(x) - 16 = 0 and use sym.fsolve
sym.solve(f - 16, x)  # find values of x which make f - 16 equal to zero
```

## Symbolic Differentiation with .diff()

```python
# define the function for this example
import sympy as sym
x=sym.symbols('x')
y = x**2 * sym.sin(1 + x**3) + sym.exp(2*x)

# derivative of y with respect to x
sym.diff(y, x)

#########################
# alternate notations:
########################
# y.diff(x)
# sym.diff(y) # <--- assumes x is the independent variable
# y.diff()    # <--- assumes x is the independent variable
```

This all works with different variables too. The code cell below is exactly like the one above, but it uses $T$ and $r$ instead of $y$ and $x$.

```python
# define the function for this example
import sympy as sym
r=sym.symbols('r')
T = r**2 * sym.sin(1 + r**3) + sym.exp(2*r)

# derivative of T with respect to r
sym.diff(T, r)

#########################
# alternate notations:
########################
# T.diff(r)
# sym.diff(T) # <--- assumes r is the independent variable
# T.diff()    # <--- assumes r is the independent variable
```

## Symbolic optimization

To identify the largest and smallest value of a contiuous function $f$ using functions from the SymPy library, perform the following steps:

1. Identify the **critical points** of $f$ which are values of in the domain of $f$ where  $f'(p) = 0$ or $f'(p)$ is undefined.
  - Apply SymPy functions `.diff()` and `.solve()` to identify critical points.
2. Compute the **critical values** $f(p)$ at each critical point in the domain.
  - Use the SymPy function `.evalf()` to compute critical values.
3. Compare the critical values with behavior of $f$ at endpoints and identify the largest and smallest values (if they exist).
  - Plot the function or use the SymPy function `.evalf()` to identify the behavior of $f$ at the endpoints.

Example: Find the largest and smallest value of $C(t)=12te^{-0.3t}$ on $0\leq t\leq 16$

```python
# load libraries and define function for this cell to run
import sympy as sym
t = sym.symbols('t')  # define symbol t as input variable
C = 12 * t * sym.exp(-0.3 * t)  # define symbolic function for C(t)

####################################
# Identify critical points of C(t)
###################################
C_prime = C.diff(t)  # find derivative of C
t_zero_derivative = sym.solve(C_prime,t) # find points where C_prime is zero
print("C'(t) = 0 at t = ", t_zero_derivative)  # print the array with all critical points

# In this case C'(t) is defined everywhere, so there is only one critical point.
# Its value is in the 1st entry of the t_zero_derivative array, so t_zero_derivative[0]

# These are the critical points and the endpoints
critical_pt = t_zero_derivative[0]
left_endpoint = 0
right_endpoint = 16

##############################################################
# Compute critical values of C(t) and values at the endpoints
##############################################################
print("Value of C(t) at critical point: ", C.evalf(subs={t: critical_pt}))
print("Value of C(t) at left endpoint: ", C.evalf(subs={t: left_endpoint}))
print("Value of C(t) at right endpoint: ", C.evalf(subs={t: right_endpoint}))

####################
# Print conclusion
####################
print("\n\nCONCLUSION:")
print("Largest value of C(t): ", C.evalf(subs={t: critical_pt}),"at t =",critical_pt)
print("Smallest value of C(t): ", C.evalf(subs={t: left_endpoint}),"at t =",left_endpoint)

# Uncomment the code below to see the corresponding plot.

# ###############################
# # plot displaying the solutions
# #############################
# critical_value = C.evalf(subs={t: critical_pt})
# t = np.linspace(0,20,1000)  # 1000 equally spaced values between t=0 and t=20
# C = 12 * t * np.exp(-0.3 * t)  # corresponding values of C
# plt.plot(t, C)  # (input, output)
# plt.vlines(x=critical_pt, ymin=0, ymax=critical_value, color='red', linestyle='dashed', linewidth=3)
# plt.xlabel('t')  # label on x-axis
# plt.ylabel('C(t)')  # label on y-axis
# plt.xlim(0, 20)  # limits on horizontal (x) axis
# plt.xticks(np.arange(0, 21, step=2))  # set tickmarks on horizontal (x) axis
# plt.ylim(0, 15)  # limits on vertical (y) axis
# plt.grid(True)  # show gridlines

# # show the plot on the screen
# plt.show()
```

## Evaluating Definite Integrals Symbolically

```python
# load simpy and declare x to be a symbolic varibale
import sympy as sym
x = sym.symbols('x')

# function to integrate and endpoints
f = sym.sin(2*x) + sym.cos(3*x)
a = 0
b = 5

# Calculate the integral of f, with respect to x, from x=a to x=b
f.integrate((x, a, b))

# alternate notation
# sym.integrate(f,(x,a,b))


# to get a decimal approximation, use float()
# float(f.integrate((x, a, b)))
```

Run the code cell below to display the signed area that the above integral corresponds to.

```python
import numpy as np
import matplotlib.pyplot as plt

# define function to integrate
def f(x):
  return np.sin(2*x) + np.cos(3*x)

# define endpoints
a = 0
b = 5

# plot
x_values = np.linspace(a, b, 200)
f_values = f(x_values)
plt.plot(x_values, f_values)
plt.fill_between(x_values, f_values, where=(x_values >= a) & (x_values <= b), alpha=0.3)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.axhline(y=0, color='black') # add a horizontal axis at y=0
plt.axvline(x=0, color='black') # add a vertical axis at x=0
plt.title('Plot of f(x) with Shaded Area Corresponding to the Integral')
plt.xlim([a-0.5, b+0.5])
plt.grid(True)
plt.show()
```

## Making a Numpy function out of a Sympy symbolic expression

This speeds up numerics considerably

```python
# define the expression for this example
import sympy as sp
import numpy as np
x=sp.symbols('x')
myExpression = x**2 * sp.sin(1 + sp.exp(2*x)) / (1 + x**3)

# there us a way to make g become a function, it uses "lambdify"
myFunction_np = sp.lambdify(x, myExpression, "numpy")

# Evaluate the numpy function ( no subs(x,3) needed )
myFunction_np(3)
```

### Make into numpy function for quick numerics


```python
import sympy as sp
x = sp.Symbol("x")
y = sp.Symbol("y")
f=(x**2+y**2)*sp.sin(x)*sp.sin(y)

# make f into a numpy function
f_np=sp.lambdify((x,y),f,"numpy")

# evaluate at x=3, y=5
f_np(3,5)
```
